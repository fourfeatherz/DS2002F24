```python
import requests
import pandas as pd

apikey = "maeundD0dq1ZlJPU0OQHCXorzXNtNtJ2TOqRKOi9"

user_input = input("Enter stock:")

url = "https://yfapi.net/v6/finance/quote"

querystring = {"symbols": user_input}

headers = {
    'x-api-key': apikey
}

response = requests.request("GET", url, headers=headers, params=querystring)
stock_json = response.json()

stocks_data = []

for stock in stock_json['quoteResponse']['result']:
    stock_data = {
        "Ticker Symbol": stock.get("symbol", "N/A"),
        "Stock Name": stock.get("longName", "N/A"),
        "Current Market Price": stock.get("regularMarketPrice", "N/A"),
        "52 Week High": stock.get("fiftyTwoWeekHigh", "N/A"),
        "52 Week Low": stock.get("fiftyTwoWeekLow", "N/A")
    }
    stocks_data.append(stock_data)

url2 = f"https://yfapi.net/v11/finance/quoteSummary/{user_input}?lang=en&region=US&modules=financialData"

querystring = {"symbols": user_input}

headers = {
    'x-api-key': apikey
}
response2 = requests.request("GET", url2, headers=headers, params=querystring)
stock_json2 = response2.json()

stock_data2 = stock_json2['quoteSummary']['result'][0]['financialData']['targetMeanPrice']['fmt']
stock_data2 = float(stock_data2)
stocks_data[0]['Target Mean Price'] = stock_data2

url3 = f"https://yfapi.net/v6/finance/recommendationsbysymbol/{user_input}"
querystring = {"symbols": user_input}

headers = {
    'x-api-key': apikey
}
response3 = requests.request("GET", url3, headers=headers, params=querystring)
stock_json3 = response3.json()
recommended_symbols = stock_json3['finance']['result'][0]['recommendedSymbols']
symbols = [stock["symbol"] for stock in recommended_symbols]


df = pd.DataFrame(stocks_data)
df['Recommended Symbols'] = ', '.join(symbols)
df.to_csv('stock_data.csv', index=False)
```
